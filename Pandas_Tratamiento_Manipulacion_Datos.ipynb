# **Problema de Negocio**

Se trata de un algoritmo capaz de analizar las características de una propiedad -como comodidades, tamaño, ocupación del espacio en un periodo determinado- y sugerir al anfitrión un precio a cobrar por tarifas diarias que garantice ganancias en momentos de alta demanda.

## **1. Entendiendo el problema**

```python
import pandas as pd
```

```python
datos = pd.read_json('datos_hosting.json')
datos.head()
```

```python
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

## **2. Datos numéricos**

```python
columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[3:])
datos
```

```python
datos.reset_index(inplace = True, drop = True)
datos.head()
```

```python
datos.info()
```

```python
import numpy as np
```

```python
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
datos.info()
```

```python
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
datos.info()
```

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos.info()
```

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
datos[['cuota_deposito','cuota_limpieza']] = datos[['cuota_deposito','cuota_limpieza']].map(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos[['cuota_deposito','cuota_limpieza']] = datos[['cuota_deposito','cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

## **3. Datos de texto**

```python
datos.head()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
datos['descripcion_local'][3169]
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r"[^a-zA-Z0-9\-\']",' ',regex=True)
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)',' ',regex=True)
```

```python
datos['descripcion_local'].head()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
datos['comodidades'] = datos['comodidades'].str.replace(r'\{|}|\"',' ',regex=True)
```

```python
datos['comodidades'] = datos['comodidades'].str.split(',')
datos.head()
```

```python
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.lower()
```

```python
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.replace(r'[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.replace(r'(?<!\w)-(?!\w)', '', regex=True)
```

```python
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.split()
datos.head()
```
